```python
import google.generativeai as genai
import os

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
genai.configure(api_key=GEMINI_API_KEY)

model = genai.GenerativeModel('models/gemini-1.5-flash-latest')

prompt = "Explique le principe de l'intelligence artificielle en une phrase simple."
response = model.generate_content(prompt)
print("Génération de texte :")
print(response.text)

prompt_autocomplete = "L'apprentissage automatique est une branche de l'intelligence artificielle qui"
response2 = model.generate_content(prompt_autocomplete)
print("\nAuto-complétion :")
print(response2.text)
```
